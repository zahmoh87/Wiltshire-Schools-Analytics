```python
def run_etl():
    import pandas as pd
    # Load the CSV file
    df = pd.read_csv('school_data.csv', encoding='latin1') # Use 'latin1' if you get encoding errors

    # Drop rows with missing school name or region
    df.dropna(subset=['SCHNAME', 'LANAME'], inplace=True)

    # Filter only active schools
    df = df[df['SCHSTATUS'] == 'Open'].copy() # Use .copy() to avoid SettingWithCopyWarning

    # Create a new column for school level
    def classify_level(row):
        if row['ISPRIMARY'] == 1:
            return 'Primary'
        elif row['ISSECONDARY'] == 1:
            return 'SECONDARY'
        elif row['ISPOST16'] == 1:
            return 'Post-16'
        else:
            return 'Other'

    # Apply the function using .loc to avoid SettingWithCopyWarning
    df.loc[:, 'School Level'] = df.apply(classify_level, axis=1)

    df.to_csv('school_cleaned.csv', index=False)
    print("ETL pipeline executed successfully.")
```

```python

import schedule
import time

# Schedule to run every Monday at 08:00

schedule.every().monday.at("08:00").do(run_etl)
while True:
  schedule.run_pending()
  time.sleep(60)
```
